```python
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500,1400,1600]
plt.plot(months, energy_consumption)
plt.xlabel('Months')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Monthly Energy Consumption')
plt.show()
```

```python
energy_sources = ['Solar', 'Wind', 'Hydropower', 'Biomass']
energy_values = [1200, 3400, 2900, 2500]
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%')
plt.title('Energy Consumption by Renewable Energy Source')
plt.show()
```

```python
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%')

plt.title('Energy Consumption by Source')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame for pair plot
data = {
    "Solar": [1200, 1500, 1300, 10],
    "Wind": [3400, 3600, 3200, 2300],
    "Hydropower": [2900, 3100, 2800, -1200],
    "Biomass": [2500, 2700, 2400, 2349]
}
df = pd.DataFrame(data)

# Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
# Create a heatmap to show the correlation matrix
correlation_matrix = df.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Between Energy Sources')
plt.show()
```

```python
# Create a box plot
sns.boxplot(data=df)

# Add title
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
# Create a violin plot
sns.violinplot(data=df)

# Add title
plt.title('Violin Plot of Energy Consumption Distributions')
plt.show()
```

```python
#histogram
df.hist(bins=2)
plt.show()
```
